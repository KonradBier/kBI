```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```
x = np.linspace(0,20,100)
y = np.random.rand(100, 4)
y = np.cumsum(y,axis = 0)
```

```
plt.plot(x, y)
```

```
sns.set()
plt.plot(x, y)
```

```
df = sns.load_dataset('tips')
df.head()
```

```
df.describe()
```

```
df.describe(include='category')
```

```
df.info()
```

```
sns.relplot(data=df, x='total_bill', y='tip')
```

```
sns.set(style='whitegrid', palette='pastel', font_scale=1)
sns.relplot(data=df, x='total_bill', y='tip', hue='size')
```

```
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='sex', sizes=(10,200), palette='viridis')
```

```
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='sex', sizes=(10,200), palette='viridis', col='time', row='smoker')
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```
df = sns.load_dataset('fmri')
df.head(3)
```

```
sns.relplot(data= df, x= 'timepoint', y= 'signal')
```

```
sns.relplot(data= df, x= 'timepoint', y= 'signal', kind='line')
```

```
df.describe()
```

```
df.timepoint.value_counts()
```

```
df.groupby('timepoint').agg('mean').plot()
```

```
sns.relplot(data= df, x= 'timepoint', y= 'signal', hue='region')
```

```
sns.relplot(data= df, x= 'timepoint', y= 'signal', hue='region', kind='line', style='event', markers=True, col='event')
```

```
sns.relplot(data= df, x= 'timepoint', y= 'signal', hue='region', kind='line', style='event', markers=True, col='subject', col_wrap=4)
```

```
import plotly.express as px
import seaborn as sns
```

```
df = sns.load_dataset('mpg')
df.head()
```

```
df.describe()
```

```
df.origin.value_counts()
```

```
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg')
```

```
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density')
```

```
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density', facet_row='origin', color='origin')
```

```
dff = df[(df.origin=='usa') | (df.origin=='europe')]
```

```
px.histogram(dff, x='mpg', color='origin', marginal='violin', title='Rozkład zmiennej mpg (USA vs. Europe)', opacity=0.7)
```

```
px.histogram(df, x='origin')
```

```
import plotly.figure_factory as ff
fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```
fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])
fig.show()
```

```
import numpy as np

x1 = np.random.randn(3000)
x2 = np.random.randn(3000) +2

ff.create_distplot([x1, x2], group_labels=['Rozkład normalny ~N(0,1)', 'Rozkład normalny ~N(2,1)'], curve_type='normal', bin_size=0.2 )
```

## **Histogram 2D**

```
import plotly.graph_objects as go
```

```
fig = go.Figure(go.Histogram2d(x=df.mpg, y=df.acceleration, nbinsx=30, nbinsy=30))
fig.add_trace(go.Scatter(x=df.mpg, y=df.acceleration, mode='markers', marker={'symbol': 'x', 'opacity':0.7, 'color': 'white', 'size':8, 'line':{'width':1}}))
fig.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```
sns.set(style='darkgrid')

df = np.random.multivariate_normal(mean=[0,1], cov=[(1, -0.4),(-0.4,1)], size=200)  #rozkład wielowymiarowy
df = pd.DataFrame(df,columns=['x','y']) # x - rozkład normalny standardowy o średniej 0 i odchyleniu 1. y - wartość oczekiwana 1, rozrzut 1
df.head(3)
```

```
df.describe()
```

```
sns.jointplot(data=df, x='x',y='y')
```

```
sns.set(style='white')
sns.jointplot(data=df, x='x',y='y', kind='hex')
```

```
sns.jointplot(data=df, x='x',y='y', kind='kde')
```

```
sns.jointplot(data=df, x='x',y='y', kind='kde', n_levels=15, cmap='viridis')
```

```
g = sns.jointplot(data=df, x='x',y='y', kind='kde', n_levels=15, cmap='viridis')
g.plot_joint(plt.scatter, c='w', s=30, marker='+')
```

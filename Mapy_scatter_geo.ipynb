```
import plotly.express as px 
```

```
df = px.data.gapminder()
df.head()
```

```
df_2007 = df[df.year==2007]
```

```
px.scatter_geo(df_2007, locations='iso_alpha', color='continent', hover_name='country', size='pop')
```

```
px.scatter_geo(df_2007, locations='iso_alpha', color='continent', hover_name='country', size='pop', projection='natural earth')
```

```
px.scatter_geo(df, locations='iso_alpha', color='continent', hover_name='country', size='pop', animation_frame='year')
```

# **choropleth map**

```
px.choropleth(df, locations='iso_alpha', color='lifeExp', hover_name='country')
```
